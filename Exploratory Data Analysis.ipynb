```python
#slicing the dataframe for the visualized correlation matrix
d = df[['duration', 'avg_vote', 'votes', 'metascore',
       'reviews_from_users', 'reviews_from_critics', 'usa_gross_income_',
       'worlwide_gross_income_', 'budget_', 'usa_revenue', 'worlwide_revenue']]
```

```python
#making a visualized correlation matrix
f = plt.figure(figsize=(8, 8))
plt.matshow(d.corr(), fignum=f.number,cmap='bone_r')
plt.xticks(range(d.select_dtypes(['number']).shape[1]), d.select_dtypes(['number']).columns, fontsize=14, rotation=60)
plt.yticks(range(d.select_dtypes(['number']).shape[1]), d.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```
